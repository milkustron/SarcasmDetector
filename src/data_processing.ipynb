```python
import pandas as pd
```

```python
data1 = pd.read_json('../data/raw/Sarcasm_Headlines_dataset_v2.json', lines = True)
data2 = pd.read_json('../data/raw/Sarcasm_Headlines_dataset.json', lines = True)
```

```python
data1
```

```python
data2 = data2[['is_sarcastic', 'headline', 'article_link']]
data2
```

```python
data.drop_duplicates(subset='headline')
```

```python
#delete article_link to avoid bias assosiated with the domain
data = data.drop(columns='article_link')
data
```
